```python
# ich versuche, zu Data umzuwandeln
data = {
    'Alter': ['26 - 35 Jahre', 'unter 25 Jahren', ''],
    'F8_Richtlinien': ['Stimme zu', 'Stimme zu', 'Trifft nicht zu'],
    'F13_Akzeptanz': ['Stimme voll zu', 'Stimme gar nicht zu', 'Trifft nicht zu']
}
df = pd.DataFrame(data)
```

```python
rename_mapping = {
    'Wie schätzen Sie Ihr technisches Wissen ein? Je mehr Sterne, umso technisch versierter.': 'F1',
    'Wie alt sind Sie?': 'F2',
    'Was ist Ihr aktueller Beschäftigungsstand? Bitte kreuzen Sie den höchsten Stand an Beschäftigung an (keine Mehrfachnennungen).\n': 'F3',
    'Was ist Ihr höchster Bildungsabschluss? Bitte kreuzen Sie den höchsten Stand an (keine Mehrfachnennungen).\n': 'F4',
    'Wie groß ist das Unternehmen, für das Sie tätig sind?': 'F5',
    'In welcher Branche ist Ihr Arbeitgeber (hauptsächlich) tätig?': 'F6',
    'Nutzen Sie im Rahmen Ihrer Arbeit IKT-Systeme (Computer, Smartphone ...)?': 'F7',
    'In meinem Unternehmen gibt es klare Richtlinien und verpflichtende Schulungen zum Umgang mit MFA.': 'F8',
    'Mein Arbeitgeber kommuniziert, warum IT-Sicherheitsmaßnahmen notwendig sind': 'F9',
    'Bei Problemen mit dem MFA-System erhalte ich Hilfe im Unternehmen': 'F10',
    'Die Konsequenzen bei Missachtung der MFA-Vorgaben sind in meinem Unternehmen klar kommuniziert.': 'F11',
    'Ihnen ist bekannt, dass der Verstoß gegen MFA-Richtlinien zu konkreten Maßnahmen geführt hat.': 'F12',
    
    'Ich finde die Verwendung von MFA sinnvoll.': 'F13',
    
    'Die berufliche Nutzung von MFA hat mich motiviert, MFA auch privat zu aktivieren.': 'F14',
    'Weil ich MFA bei der Arbeit nutze, verstehe ich die Technologie generell besser.': 'F15',
    'Ich würde MFA privat nicht nutzen, wenn ich die Nutzung nicht bei der Arbeit gezeigt bekommen hätte.': 'F16',
    'Ich nutze MFA privat nur bei Diensten, bei denen ich dazu gezwungen werde.': 'F17',
    
    "Zur Sicherstellung der Datenqualität wählen Sie hier bitte die Option 'Stimme voll zu' aus.": 'F18',
    'Haben Sie Anmerkungen zum Thema Multifaktor-Authentifizierung (MFA) oder gibt es Aspekte Ihres Nutzungsverhaltens (beruflich oder privat), die in den vorherigen Fragen nicht berücksichtigt wurden?': 'F19'
}
```

```python
def encode_likert(val):
    mapping = {
        'Stimme gar nicht zu': 1,
        'Stimme nicht zu': 2,
        'Neutral': 3,
        'Stimme zu': 4,
        'Stimme voll zu': 5,
        'Trifft nicht zu': 0
    }
    return mapping.get(val, 3)

df['F8_num'] = df['F8_Richtlinien'].apply(encode_likert)
df['F13_num'] = df['F13_Akzeptanz'].apply(encode_likert)

print(df[['F8_num', 'F13_num']].head())
```

```python
df_mapped = df.rename(columns=rename_mapping)
print(df_mapped.head())
```
